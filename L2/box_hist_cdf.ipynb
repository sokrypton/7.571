```
def calc_ecdf(input_values):
  x_values = np.sort(input_values) #this just sorts our cdk2 values, and sets them as the x_values for our plot.
  total_values = x_values.shape[0] #this gives us the total number of points
  y_values = np.arange(1, total_values+1)/float(total_values) #this is a list from 1 to the number of points, with each point divided by the total number of points
  return x_values, y_values
```

```
hx, hy = calc_ecdf(cdk2_levels['healthy'])
```

```
a = plt.plot(hx, hy)
```

```
ux,uy = calc_ecdf(cdk2_levels['unknown'])
```

```
a = plt.plot(hx, hy)
a = plt.plot(ux, uy)
```

```
tx, ty = calc_ecdf(cdk2_levels['tumor'])
a = plt.plot(hx, hy)
a = plt.plot(ux, uy)
a = plt.plot(tx, ty)
```
